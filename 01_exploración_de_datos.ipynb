```python
# Librerías necesarias
import os
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import random
```

```python
data_path = "/content/tomato-leaf-dataset/tomato/train"

# Ver clases disponibles
clases = sorted(os.listdir(data_path))
print("Número de clases:", len(clases))
print("Clases:", clases)
```

```python
# Contar imágenes por clase
conteo = {}
for clase in clases:
    ruta_clase = os.path.join(data_path, clase)
    imagenes = [f for f in os.listdir(ruta_clase) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    conteo[clase] = len(imagenes)

# Mostrar conteo
for clase, cantidad in conteo.items():
    print(f"{clase}: {cantidad} imágenes")
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x=list(conteo.keys()), y=list(conteo.values()))
plt.title("Número de imágenes por clase")
plt.xlabel("Clase")
plt.ylabel("Cantidad de imágenes")
plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(18, 7))
fig.suptitle("Ejemplos de imágenes de hojas de tomate", fontsize=16)

for i, clase in enumerate(random.sample(clases, 5)):
    ruta_clase = os.path.join(data_path, clase)
    imagen_path = os.path.join(ruta_clase, random.choice(os.listdir(ruta_clase)))
    imagen = Image.open(imagen_path)

    axes[0, i].imshow(imagen)
    axes[0, i].axis('off')
    axes[0, i].set_title(clase, fontsize=10)

for i, clase in enumerate(random.sample(clases, 5)):
    ruta_clase = os.path.join(data_path, clase)
    imagen_path = os.path.join(ruta_clase, random.choice(os.listdir(ruta_clase)))
    imagen = Image.open(imagen_path)

    axes[1, i].imshow(imagen)
    axes[1, i].axis('off')
    axes[1, i].set_title(clase, fontsize=10)

plt.tight_layout()
plt.show()
```
